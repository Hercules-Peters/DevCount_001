```python
import pandas as pd
import yfinance as yf
from datetime import date, timedelta
```

```python
def download_clean_data(start_date, end_date):
    # Download data from Yahoo Finance
    data = yf.download('WLD-USD',
                       start=start_date,
                       end=end_date,
                       progress=False)

    # Reset index and rename columns
    data.reset_index(inplace=True)
    data.rename(columns={'Date': 'Date', 'Open': 'Open', 'High': 'High', 'Low': 'Low', 'Close': 'Close', 'Adj Close': 'Adj_Close', 'Volume': 'Volume'}, inplace=True)

    return data
```

```python
# Define start and end dates for data retrieval
today = date.today()
end_date = today.strftime("%Y-%m-%d")
start_date = (today - timedelta(days=258)).strftime("%Y-%m-%d")

# Download and clean the data
data = download_clean_data(start_date, end_date)

# Save data to a CSV file
data.to_csv('WLD_data.csv', index=False)

# Extract features and target variable
X = data[['Open', 'High', 'Low', 'Adj_Close', 'Volume']]
y = data['Close']
```

```python
import plotly.graph_objects as go
figure = go.Figure(data=[go.Candlestick(x=data["Date"],
                                        open=data["Open"],
                                        high=data["High"],
                                        low=data["Low"],
                                        close=data["Close"])])
figure.update_layout(title="Worldcoin Price Analysis",
                     xaxis_rangeslider_visible=False)
figure.show()
```
